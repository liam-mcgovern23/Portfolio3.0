**DATA CLEANING AND PREP**

```python
# Check for missing values and data types
missing_data = df.isnull().sum()
data_types = df.dtypes

missing_data, data_types
```

```python
# Step 1: Remove unnecessary columns
data_cleaned = df.drop(columns=['Unnamed: 0', 'Matches'])
```

```python
# Step 2: Handle missing values
# Fill numeric columns with missing values with 0 (assuming no activity recorded)
numeric_columns = data_cleaned.select_dtypes(include=['float64']).columns
data_cleaned[numeric_columns] = data_cleaned[numeric_columns].fillna(0)
```

```python
# Step 3: Convert data types
data_cleaned['Age'] = data_cleaned['Age'].astype(int)
data_cleaned['Born'] = data_cleaned['Born'].astype(int)
```

```python
# Step 4: Check for and remove any duplicate entries
data_cleaned = data_cleaned.drop_duplicates()
```

```python
# Display the cleaned data info and a few rows to confirm changes
cleaned_data_info = data_cleaned.info()
cleaned_data_head = data_cleaned.head()

cleaned_data_info, cleaned_data_head
```

**CLUSTERING ANALYSIS**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Select features for clustering
features_for_clustering = data_cleaned[['Gls', 'Sh', 'SoT', 'xG', 'Sh/90', 'SoT/90']]
# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_for_clustering)

# Find the optimal number of clusters using the silhouette score
silhouette_scores = []
for k in range(2, 11):  # Test between 2 and 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    score = silhouette_score(features_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

**T-TEST**

```python
from scipy.stats import ttest_ind

# Filter the dataset for defenders and forwards
defenders_goals = data_cleaned[data_cleaned['Pos'].str.contains('DF')]['Gls']
forwards_goals = data_cleaned[data_cleaned['Pos'].str.contains('FW')]['Gls']

# Perform a T-test between the goal scores of defenders and forwards
t_test_results = ttest_ind(defenders_goals, forwards_goals, equal_var=False)  # Using Welch's t-test for possibly unequal variances

t_test_results
```

**ANOVA**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Prepare data for ANOVA by creating a simplified position category (ignoring combined positions like 'FW,MF')
data_cleaned['Simple_Pos'] = data_cleaned['Pos'].apply(lambda x: 'FW' if 'FW' in x else ('MF' if 'MF' in x else ('DF' if 'DF' in x else 'GK')))

# ANOVA test
anova_model = ols('Gls ~ C(Simple_Pos)', data=data_cleaned).fit()
anova_results = sm.stats.anova_lm(anova_model, typ=2)  # Type 2 ANOVA DataFrame

anova_results
```

**LINEAR REGRESSION**

```python
import statsmodels.api as sm

# Setting up the independent variables (predictors) and the dependent variable (response)
X = data_cleaned[['Sh', 'SoT', 'xG', 'Sh/90', 'SoT/90']]
y = data_cleaned['Gls']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Fitting the linear regression model
model = sm.OLS(y, X).fit()

# Getting the summary of the regression model
model_summary = model.summary()
model_summary
```

**LOGISTIC REGRESSION**

```python
from statsmodels.discrete.discrete_model import Logit

# Create a binary outcome variable for scoring at least one goal
data_cleaned['Scored'] = (data_cleaned['Gls'] >= 1).astype(int)

# Setting up the independent variables (predictors) and the dependent variable (response)
X_logistic = data_cleaned[['Sh', 'SoT', 'xG', 'Sh/90', 'SoT/90']]
y_logistic = data_cleaned['Scored']

# Adding a constant to the model (intercept)
X_logistic = sm.add_constant(X_logistic)

# Fitting the logistic regression model
logistic_model = Logit(y_logistic, X_logistic).fit()

# Getting the summary of the logistic regression model
logistic_model_summary = logistic_model.summary()
logistic_model_summary
```

**RANDOM FORESTFOR SCORING PREDICTIONS**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Features and target
X_rf = data_cleaned[['Sh', 'SoT', 'xG', 'Sh/90', 'SoT/90']]
y_rf = data_cleaned['Scored']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Predict on the testing set
y_pred = rf_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

accuracy, classification_rep
```

**FEATURE IMPORTANCE**

```python
from sklearn.inspection import permutation_importance

# Calculate feature importance
feature_importances = rf_classifier.feature_importances_

# Also perform permutation importance which is often more reliable as it accounts for feature interactions
perm_importance = permutation_importance(rf_classifier, X_test, y_test, n_repeats=10, random_state=42)

# Displaying feature importance
feature_imp_dict = {
    'Feature': X_train.columns,
    'Importance': feature_importances,
    'Permutation Importance Mean': perm_importance.importances_mean,
    'Permutation Importance Std Dev': perm_importance.importances_std
}

feature_importance_df = pd.DataFrame(feature_imp_dict)
feature_importance_df.sort_values(by='Permutation Importance Mean', ascending=False)
```

**TUNING**

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30]
}

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=5,  # 5-fold cross-validation
                           scoring='accuracy',  # scoring metric
                           verbose=1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_params, best_score
```

**SUPPORT VECTOR MACHINE**
